```python
from sentence_transformers import SentenceTransformer
import faiss

dimension = 768
model = SentenceTransformer('pritamdeka/S-BioBert-snli-multinli-stsb')
query_index = faiss.IndexFlatL2(dimension)
```

```python
import numpy as np
def cos_sim(e1, e2):
    return np.dot(e1, e2) / (np.linalg.norm(e1) * np.linalg.norm(e2))
```

```python
query_dict = []

question = ""
answer = ""

embedding = model.encode(question, convert_to_tensor=True)
embedding = embedding.cpu().numpy().astype('float32')
embedding = embedding.reshape(1, -1)

new_entry = {
    "question": question,
    "embedding": embedding,
    "answer": answer
}
```

```python
def cache_hit(new_query, index, q_dict):
  new_query_embedding = model.encode(new_query, convert_to_tensor=True)
  new_query_embedding = new_query_embedding.cpu().numpy().astype('float32')
  new_query_embedding = new_query_embedding.reshape(1, -1)

  for cache in q_dict:
    if cos_sim(new_query_embedding, cache["embedding"]) > 0.9:
      return cache["answer"]

  return None
```
